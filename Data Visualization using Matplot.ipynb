```python
import pandas as pd
 
url = "https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Stock/MSFT_data.csv"
```

```python
df=pd.read_csv(url)

df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(25,10)) #w , h

df1=df.head(60)

plt.plot(df1['date'],df1['open'])
plt.show()
```

```python
plt.figure(figsize=(25,10)) #w , h

df1=df.head(60)

plt.plot(df1['date'],df1['open'])
plt.xlabel('Date')
plt.ylabel('Opening price of MST Stock')
plt.title('Date vs Open Price')
plt.show()
```

```python
plt.figure(figsize=(25,10)) #w , h

df1=df.head(60)

plt.plot(df1['date'],df1['open'])
plt.xlabel('Date')
plt.ylabel('Opening price of MST Stock')
plt.title('Date vs Open Price')

plt.xticks(df1['date'],rotation=90)
plt.show()
```

```python
x=np.arange(-55,55)
plt.plot(x, x**2)
plt.plot(x,x**3)
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
```

```python
x=np.arange(-5,15)
plt.plot(x, x**2, c='black',marker ='>')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X** 2', 'Y =X ** 3'])
plt.show()
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X** 2', 'Y =X ** 3'],loc='upper center')
plt.show()
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X** 2', 'Y =X ** 3'],loc='upper center')
plt.text(1,1,"(1,1)")
plt.text(15,55,"(15,55)")

plt.show()
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X** 2', 'Y =X ** 3'],loc='upper center')
plt.text(1,1,"(1,1)")
plt.text(15,55,"(15,55)")
plt.grid()


plt.show()
```

```python
x=np.arange(-5,10)
plt.plot(x, x**2, c='black',marker ='>',linestyle='-.')
plt.plot(x,x**3 ,'c',marker='*')
plt.xlabel("x")
plt.ylabel("Y=f(x)")
plt.title("Y=F(X)")
plt.legend(['Y = X** 2', 'Y =X ** 3'],loc='upper center')
plt.text(1,1,"(1,1)")
plt.text(15,55,"(15,55)")
plt.grid()
plt.axis("off")

plt.show()
```

### Vectorisation

```python
x=[1,2,3,4,5]
y=np.array([2,4,5,6,8])
y2=-y
y3=[5,5,5,5,5]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y3)
plt.show()
```

```python
plt.scatter(y,y3)
plt.show()
```

```python
plt.scatter(y,y2)
plt.show()
```

```python
plt.scatter(y,x,c=['b','r','g','black','#ffaabb'])
plt.show()
```

```python
plt.scatter(y,x,marker='>',c=['b','r','g','black','#ffaabb'],alpha=0.6,s=y**3)
plt.show()
```

```python
plt.scatter(y,x,marker='>',c=['b','r','g','black','#ffaabb'],alpha=0.6,s=y**3,edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.scatter(y,x,marker='>',c=['b','r','g','black','#ffaabb'],alpha=0.6,s=y**3,edgecolor='black')
plt.show()
```
