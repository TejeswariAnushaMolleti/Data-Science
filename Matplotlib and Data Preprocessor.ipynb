```python
import matplotlib.pyplot as plt
```

```python
x=['x','y','z','a','b']
y=[12,37,89,45,67]

plt.bar(x,y,color=['c','r','b','y','#00ffbb'],width=0.5)
plt.title("Bar Plot")
plt.xlabel('names')
plt.ylabel('marks')
plt.show()
```

```python
x=['x','y','z','a','b']
y=[12,37,89,45,67]

plt.barh(x,y,color=['c','r','b','y','#00ffbb'])
plt.text(12,'x','pandas')
plt.text(37,'y','numpy')
plt.title("Bar Plot")
plt.xlabel('names')
plt.ylabel('marks')
plt.show()
```

## Pie plot
- matplot.ib.pyplot.pie

```python
marks=[12,34,56,67,78]
plt.pie(marks,labels=['a','b','c','d','e'])
```

```python
marks=[12,34,56,67,78]
plt.pie(marks,labels=['a','b','c','d','e'],explode=[0,0.25,0.1,0.01,0.05])
```

```python
marks=[12,34,56,67,78]
plt.pie(marks,labels=['a','b','c','d','e'],explode=[0,0.25,0.1,0.01,0.05],labeldistance=0.5)
```

```python
marks=[12,34,56,67,78]
n=[i for i in range (len(marks))]
plt.pie(marks,labels=n)
```

```python
marks=[12,34,56,67,78]
plt.pie(marks,labels=marks,radius=1.5,labeldistance=0.5,colors=['c','g','y','#00ffab','m'])
plt.show() #remoe unneccasary data on the top
```

## Standard Scalar
- mean =0 and std=1
    z=x-mue/standarddev

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
s
```

```python
std_data=s.fit_transform(df)
std_data #df data to standard dist of data
```

```python
std1=pd.DataFrame(std_data,columns=df.columns)
std1.head()
```

```python
std1.describe()
```

```python
import seaborn as sns
plt.title("After applying scaling")
sns.kdeplot(std1['TV'])
sns.kdeplot(std1['radio'])
sns.kdeplot(std1['newspaper'])
sns.kdeplot(std1['sales'])
```

## Minmax Scalar

```python
from sklearn.preprocessing import MinMaxScaler
s= MinMaxScaler()
minmaxd=s.fit_transform(df)
minmaxd
```
