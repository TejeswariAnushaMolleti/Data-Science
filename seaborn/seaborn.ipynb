##  Creation of subplots
- plt.subplot()

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1) #rows,cols,index/position
plt.plot([1,2,3],[2,3,4])
#index pos must be < rows n cols
plt.subplot(1,2,2)
plt.scatter([1,2,3,4,5],[5,6,7,8,9])
```

```python
import numpy as np
plt.subplot(2,1,1) 
plt.bar(['x','y','z'],height=6)
plt.subplot(2,1,2) 
plt.hist(np.random.randint(1,10,5),bins=10)
```

```python
x=np.random.randint(1,10,5)
y=np.random.randint(1,100,10)
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.plot(x,x,'r*',label="line plot")
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.scatter(x,y,label="scatter plot")
plt.plot(x,x,'r*',label="line")
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9])
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.scatter(x,y,label="scatter plot")
plt.legend()
plt.subplot(2,2,2)
plt.plot(x,x,'r',label="line")
plt.legend()
plt.subplot(2,2,3)
plt.bar(x,x,label="line")
plt.legend()
plt.subplot(2,2,4)
plt.barh(x,x,label="line")
plt.legend('upper right')
```

```python
plt.subplot(1,2,1)
plt.pie([20,30,40],labels=['python','java','R'])
plt.subplot(1,2,2)
plt.barh([20,30,40],width=5,color=['c','b','m'],label='bar graph')
plt.legend()
plt.text(-1.5,20,"m1")
plt.text(5,30,"m2")
plt.text(-1.5,40,"m3")
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(),'m')
ax2=fig.add_subplot(2,2,2)
ax2.hist(np.random.randn(10),bins=20,color='y',orientation='horizontal')
ax3=fig.add_subplot(2,2,3)
ax3.scatter(np.arange(10),np.random.randint(1,30,10),color='k',label='scattet')
ax3.legend()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(),'k',label='line plot')
plt.legend()
ax2=fig.add_subplot(2,2,2)
ax2.hist(np.random.randn(100),bins=20,color='y',orientation='horizontal',label="hist plot")
ax3=fig.add_subplot(2,2,3)
ax3.scatter(np.arange(30),np.random.randint(1,50,30),label='scatter plot')
plt.show()
```
